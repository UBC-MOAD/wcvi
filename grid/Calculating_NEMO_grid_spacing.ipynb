```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import matplotlib as mpl
import os,sys
import numpy as np
from salishsea_tools import viz_tools, geo_tools,nc_tools
from grid_alignment import calculate_initial_compass_bearing as cibc
from scipy.interpolate import griddata, interp1d
import matplotlib.cm as cm

%matplotlib inline
```

```python
coord = nc.Dataset('/ocean/ssahu/CANYONS/wcvi/grid/coordinates.nc')
```

```python
nc_tools.show_variables(coord)
```

```python
e1t = coord.variables['e1t'][0,...];
e2t = coord.variables['e2t'][0,...];

e1u = coord.variables['e1u'][0,...];
e2u = coord.variables['e2u'][0,...];

print(e1t.shape, e2t.shape)
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2,figsize=(10,10))

mesh = ax0.pcolormesh(e1t, cmap =cm.Spectral_r)
viz_tools.set_aspect(ax0)
fig.colorbar(mesh,ax=ax0)

mesh = ax1.pcolormesh(e2t, cmap =cm.Spectral_r)
viz_tools.set_aspect(ax1)
fig.colorbar(mesh,ax=ax1)

plt.draw()
plt.show()
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2,figsize=(10,10))

mesh = ax0.pcolormesh(e1u, cmap =cm.Spectral_r)
viz_tools.set_aspect(ax0)
fig.colorbar(mesh,ax=ax0)

mesh = ax1.pcolormesh(e2u, cmap =cm.Spectral_r)
viz_tools.set_aspect(ax1)
fig.colorbar(mesh,ax=ax1)

plt.draw()
plt.show()
```

```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
import matplotlib as mpl
from salishsea_tools import viz_tools, geo_tools,nc_tools
```

```python
fname_V_43 = '/home/ssahu/saurav/JP_BC/cat_43_V.nc'
```

```python
nc_tools.show_variables(nc.Dataset(fname_V_43))
nc_tools.show_dimensions(nc.Dataset(fname_V_43))
```

```python
with nc.Dataset(fname_V_43, 'r') as cncat_v_43:
    V_43_unfiltered = cncat_v_43.variables['vomecrty'][:];
    time_JP_43 =  cncat_v_43.variables['time_counter'][:]; 

    
    
print(V_43_unfiltered.shape, time_JP_43.shape)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(25, 10))

plt.plot(time_JP_43, V_43_unfiltered[:,0,530,250],'-b')

plt.plot(time_JP_43, V_43_unfiltered[:,0,570,250],'-r')

plt.plot(time_JP_43, V_43_unfiltered[:,0,600,250],'-y')

plt.xlabel('Time (daily) for 43 days')

plt.ylabel('V (in m/sec)')

plt.title('Red= shelf break, blue = 2000m, yellow = shelf')

plt.show()
```

```python
V_43_filtered = V_43_unfiltered
for idx, val in enumerate(time_JP_43):
    V_43_filtered[idx-1,...] = V_43_unfiltered[idx-2,...]*0.25 + V_43_unfiltered[idx-1,...]*0.5 + \
                                V_43_unfiltered[idx,...]*0.25
    
print(V_43_filtered.shape)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(25, 10))

plt.plot(time_JP_43, V_43_filtered[:,0,530,250],'-b')

plt.plot(time_JP_43, V_43_filtered[:,0,570,250],'-r')

plt.plot(time_JP_43, V_43_filtered[:,0,600,250],'-y')

plt.xlabel('Time (daily) for 43 days')

plt.ylabel('V (in m/sec)')

plt.title('Red= shelf break, blue = 2000m, yellow = shelf')

plt.show()
```
